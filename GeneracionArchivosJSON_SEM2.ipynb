```python
import json
import torch
import flash
from flash.audio import SpeechRecognition, SpeechRecognitionData
```

#TIC TAC TOE

```python
#Los Links que salen tanto en columnas como filas varian del drive de la persona
columnas = ["/content/drive/MyDrive/OctavoCiclo/TDP/speechrecognition/CertificationTest/a.wav","/content/drive/MyDrive/OctavoCiclo/TDP/speechrecognition/CertificationTest/be.wav","/content/drive/MyDrive/OctavoCiclo/TDP/speechrecognition/CertificationTest/ce.wav"]
index = 0

data_json = {"columna": [], "fila": []}

model = SpeechRecognition.load_from_checkpoint("/content/drive/MyDrive/OctavoCiclo/TDP/modelo/Modelo_30ep_thirdTrain_part2.pt")
trainer = flash.Trainer()

for k in columnas:
  datamodule = SpeechRecognitionData.from_files(predict_files=[k], batch_size=4)
  predictions = trainer.predict(model, datamodule=datamodule)
  data_json["columna"].append({"key":predictions[0][0],"value":index})
  index = index + 1

#Los Links que salen tanto en columnas como filas varian del drive de la persona
filas = ["/content/drive/MyDrive/OctavoCiclo/TDP/speechrecognition/CertificationTest/uno.wav","/content/drive/MyDrive/OctavoCiclo/TDP/speechrecognition/CertificationTest/dos.wav","/content/drive/MyDrive/OctavoCiclo/TDP/speechrecognition/CertificationTest/tres.wav"]
index = 0
for k in filas:
	datamodule = SpeechRecognitionData.from_files(predict_files=[k], batch_size=4)
	predictions = trainer.predict(model, datamodule=datamodule)
	data_json["fila"].append({"key":predictions[0][0],"value":index})
	index = index + 3

json_object = json.dumps(data_json, indent = 4) 
with open("/content/drive/MyDrive/OctavoCiclo/JSONTICTACTOE.txt", "w") as f:
	f.write(json_object)
```
