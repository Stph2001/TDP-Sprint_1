## Importando librerías

```python
import torch
import flash
from flash.audio import SpeechRecognition, SpeechRecognitionData
```

## Realizando Fine Tuning

```python
datamodule = SpeechRecognitionData.from_csv(
    "file",
    "text",
    train_file="/content/drive/MyDrive/TDP/Sprint1/dataset.csv",
    test_file="/content/drive/MyDrive/TDP/Sprint1/dataset.csv",
    batch_size=4
)
```

```python
model = SpeechRecognition.load_from_checkpoint("/content/drive/MyDrive/TDP/Sprint1/Modelo_30ep_thirdTrain_part2.pt")
```

```python
trainer = flash.Trainer(max_epochs=30, gpus=torch.cuda.device_count())
trainer.finetune(model, datamodule=datamodule, strategy="freeze")
```

## Guardando el nuevo Modelo

```python
trainer.save_checkpoint("/content/drive/MyDrive/TDP/Sprint1/Modelo_Entrenado.pt")
```
