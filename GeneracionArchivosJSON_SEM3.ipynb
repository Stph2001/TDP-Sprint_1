```python
import json
import torch
import flash
from flash.audio import SpeechRecognition, SpeechRecognitionData
```

```python
print(predictions)
```

#TIC TAC TOE

```python
index = 0

data_json = {"columna": [], "fila": []}

for i in range(3):
  data_json["columna"].append({"key":predictions[i][0],"value":index})
  index = index + 1

index = 0
for i in range(11,14):
	data_json["fila"].append({"key":predictions[i][0],"value":index})
	index = index + 3

json_object = json.dumps(data_json, indent = 4) 
with open("/content/drive/MyDrive/OctavoCiclo/JSONTICTACTOE.json", "w") as f:
	f.write(json_object)
```

#DAMAS

```python
index = 0

data_json = {"columna": [], "fila": []}

for i in range(3,11):
	data_json["columna"].append({"key":predictions[i][0],"value":index})
	index = index + 1

index = 7

for i in range(11,19):
	data_json["fila"].append({"key":predictions[i][0],"value":index})
	index = index - 1

json_object = json.dumps(data_json, indent = 4) 
with open("/content/drive/MyDrive/OctavoCiclo/JSONDAMAS.json", "w") as f:
	f.write(json_object)
```
